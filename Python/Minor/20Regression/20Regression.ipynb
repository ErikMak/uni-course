Загрузите датасет hostels.csv по ссылке. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

df = pd.read_csv('hostels.csv')
df.info()
```

Создайте датафрейм X, исключив столбец Рейтинг. Этот датафрейм будет содержать признаки для обучения модели (оценки хостелов посетителями сайтов Booking.com и TripAdvisor.com).

```python
X = df.drop(columns=['Рейтинг'])
```

Запишите данные из столбца Рейтинг в масссив y. Это целевая переменная – рейтинг хостела.

```python
y = df['Рейтинг']
```

Разбейте Х и y на две части. Одна – для обучения модели (X_train), а другая – для проверки точности предсказания модели (X_test). В функции train_test_split задайте параметры test_size=0.3 и random_state=1.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# DecisionTree
Обучите модель дерева решений на данных из X_train. Укажите параметр random_state=1.

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=1)
model.fit(X_train, y_train)
```

Получите предсказания для данных, на которых обучали (X_train), и для тестовых данных (X_test) и найдите значение метрики R2 (коэффициент детерминации) для трейна и теста с точность три знака после запятой.

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Вычисление метрики R2 для трейна и теста
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Вывод результатов с точностью до трех знаков после запятой
print(f'R2 train: {r2_train:.3f}')
print(f'R2 test: {r2_test:.3f}')
```

# RandomForest
Обучите модель случайного леса на данных из X_train. Укажите параметр random_state=1 и число деревьев n_estimators=100.

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=1)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Вычисление метрики R2 для трейна и теста
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Вывод результатов с точностью до трех знаков после запятой
print(f'R2 train: {r2_train:.3f}')
print(f'R2 test: {r2_test:.3f}')
```
