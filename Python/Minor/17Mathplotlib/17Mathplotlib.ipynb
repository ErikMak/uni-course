Загрузите датасет museum_visitors.csv и постройте графики посещаемости музеев. 

Какому музею соответствует график?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('museum_visitors.csv')
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

plt.figure(figsize=(14, 8))

for column in data.columns:
    plt.plot(data.index, data[column], label=column)

plt.title('Посещаемость музеев')
plt.xlabel('Дата')
plt.ylabel('Количество посетителей')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

Загрузите датасет flight_delays.csv с данными о задержках рейсов различных авиакомпаний за год.

На графике представлены распределения для сравнения задержек рейсов двух авиакомпаний. Определите эту пару авиакомпаний.

```python
data = pd.read_csv('flight_delays.csv')
data.info()
```

```python
data['DL'] = data['DL'].astype(float)
data['UA'] = data['UA'].astype(float)

# Построение графика распределения задержек для двух авиакомпаний
plt.figure(figsize=(12, 6))

# Используем seaborn для построения графиков распределения
sns.kdeplot(data['WN'], label='Delta Air Lines (DL)', fill=True, alpha=0.5)
sns.kdeplot(data['NK'], label='United Airlines (UA)', fill=True, alpha=0.5)

plt.title('Сравнение задержек рейсов: Delta Air Lines vs United Airlines')
plt.xlabel('Задержка (минуты)')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
